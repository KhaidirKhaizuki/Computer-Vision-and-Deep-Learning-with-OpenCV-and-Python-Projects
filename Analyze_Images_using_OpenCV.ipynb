```python
#Download the image.
cv2.imwrite("HuaweiEdges.png", cv2.cvtColor(inverted_image, cv2.COLOR_RGB2BGR) )
```

# Detecting Corners in Images.

```python
image = cv2.imread("Huawei-Logo.png")
pyplot.rcParams["figure.figsize"] = (5,5)
pyplot.imshow(image)
```

```python
image = cv2.resize(image,(0,0),fx=0.75,fy=0.75)
cv2_imshow(image)
```

```python
#convert grayscal images
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray_image)
```

```python
#this function will detect corner into a list of multidimensional array  of coordinates
MAXIMUM_CORNERS = 100
QUALITY_LEVEL = 0.01 
MINIMUM_DISTANCE = 10

corners = cv2.goodFeaturesToTrack(gray_image,MAXIMUM_CORNERS,QUALITY_LEVEL, MINIMUM_DISTANCE )

corners
```

```python
#convert into int values
import numpy

corners = numpy.int0(corners)

corners
```

```python
RADIUS = 1

COLOR = (0,255,0)

for corner in corners:

   x , y = corner.ravel()

   cv2.circle(image, (x,y), RADIUS,  COLOR, -1)

cv2_imshow(image) 
```

```python
#Download the image.
cv2.imwrite("HuaweicORNER.png", image )
```

# Detect Contours in Image


```python
image = cv2.imread("Huawei-Logo.png")
pyplot.rcParams["figure.figsize"] = (5,5)
pyplot.imshow(image)
```

```python
#convert grayscal images
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray_image)
```

```python
pyplot.contour(gray_image,  origin ="image")
```

```python
ret, thresh = cv2.threshold(gray_image, 150, 255, 0 )


contours , hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(image,contours,-1, (0,255,0), 3 )

pyplot.imshow(image)
```
